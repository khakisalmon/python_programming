```python
# 1. 2017년 5월과 비교하여 2023년 5월 시점에서 추가된 도서의 목록 및 개수

import pandas as pd

def get_added_books(file1, file2):
    # 파일 1 읽기
    df1 = pd.read_csv(file1, encoding='euc-kr', error_bad_lines=False)


    # 파일 2 읽기
    df2 = pd.read_csv(file2, encoding='euc-kr', error_bad_lines=False)

    # 추가된 도서 목록
    added_books = []

    # 파일 1에는 있지만 파일 2에는 없는 도서를 찾아 추가된 도서 목록에 추가
    added_books = df1[~df1['도서명'].isin(df2['도서명'])]['도서명'].tolist()

    # 추가된 도서 개수
    added_count = len(added_books)

    return added_books, added_count

# 대출 목록 파일 경로
file_2023_may = "성동구립성수도서관 장서 대출목록 (2023년 05월).csv"
file_2017_may = "성동구립성수도서관 장서 대출목록 (2017년 05월).csv"

# 추가된 도서 목록 및 개수 가져오기
added_books, added_count = get_added_books(file_2023_may, file_2017_may)

# 결과 출력


print("추가된 도서 목록:")
for book in added_books:
    print(book)

print(f"추가된 도서 개수: {added_count}")
```

```python
import pandas as pd



def get_removed_books(file1, file2):
    # Read file 1
    df1 = pd.read_csv(file1, encoding='euc-kr', error_bad_lines=False)

    # Read file 2
    df2 = pd.read_csv(file2, encoding='euc-kr', error_bad_lines=False)
 
    removed_books = []



    # Get removed books list
    removed_books = df2[~df2['도서명'].isin(df1['도서명'])]['도서명'].tolist()

    # Get count of removed books
    removed_count = len(removed_books)

    return removed_books, removed_count

file_2023_may = "성동구립성수도서관 장서 대출목록 (2023년 05월).csv"
file_2017_may = "성동구립성수도서관 장서 대출목록 (2017년 05월).csv"

# Get the list and count of removed books
removed_books, removed_count = get_removed_books(file_2023_may, file_2017_may)

# Print the results


print("삭제된 도서 목록:")
for book in removed_books:
    print(book)

print(f"삭제된 도서 개수: {removed_count}")
```
